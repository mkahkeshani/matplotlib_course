```python
import matplotlib.pyplot as plt
import numpy as np
```

### نمودار میله ای ساده

```python
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021,]
income = [82000, 75000, 79000, 63000, 68000, 72000, 82300,]
expenses = [80000, 70000, 78000, 60000,59000, 51000, 74000,]

avr = sum(years) / len(income)

plt.bar(years,income, width = 0.5, color = 'c' , edgecolor = 'black', bottom = avr)
plt.show()
```

### نمودارمیله ای گروهی

```python
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021,]
income = [82000, 75000, 79000, 63000, 68000, 72000, 82300,]
expenses = [80000, 70000, 78000, 60000,59000, 51000, 74000,]

x = np.arange(len(years))
wid = 0.4

plt.bar(x,income, width = wid)
plt.bar(x - wid ,expenses, color = 'r', width = wid )
plt.xticks(x , years)

#plt.plot(years,expenses, color = 'r' )
plt.show()
```
