```python
import matplotlib.pyplot as plt
import pandas as pd
```

# رسم نمودار جعبه برای نمایش توزیع طول گلبرگ ها در سه نمونه گل زنبق

```python
# خواندن و نمایش دیتاست
data = pd.read_csv('iris.csv')
data
```

```python
# گروه بندی داد ها بر اساس گونه
# مقادیر منحصر به فرد هر گونه را استخراج می کند و در یک آرایه ریخته و به صورت لیست برگردانده می شوند
species = data['species'].unique()

# حلقه برای هر کدام نمونه از نمونه های موجود     
# در ستون گونه ها اجرا می شود و با هر بار اجرا یکی از گونه هار گرفته و در یک سری جدید ذخیره می کند
# در نهایت یک دیتا فریم جدید ساخته می شود که شامل ردیف هایی از مقادیر اصلی است و داده ها را بر اساس هر گونه فیلتر می کند
data_to_plot = [data[data['species'] == sp]['petal_length'] for sp in species]


# ایجاد یک فیگور و یک محور
fig, ax = plt.subplots(figsize=(8, 5))

# رسم نمودار جعب ای
ax.boxplot(data_to_plot, labels=species, notch= True, vert = True, patch_artist = False, showmeans = True
          ,showcaps = True, showbox = True, whis = 1.5)

# تنظیم عنوان و برچس های محورها
ax.set_title('distribution of petal length in three different species')
ax.set_xlabel('species')
ax.set_ylabel('petal_length')

# نمایش نمودار
plt.show()
```
