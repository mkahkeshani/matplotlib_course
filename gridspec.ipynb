```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as grid
```

```python
fig, ax = plt.subplots(ncols = 2, nrows = 2, facecolor = 'b')
```

```python
fig = plt.figure(constrained_layout = True, facecolor = 'r')
spec = grid.GridSpec(ncols = 2, nrows = 2, figure=fig)

ax1 = fig.add_subplot(spec[0,0])
ax2 = fig.add_subplot(spec[0,1])
ax3 = fig.add_subplot(spec[1,0])
ax1 = fig.add_subplot(spec[1,1])
```

```python
fig = plt.figure(constrained_layout = True, facecolor = 'y')
grid = fig.add_gridspec(3,3)

ax1 = fig.add_subplot(grid[0, :])
ax2 = fig.add_subplot(grid[1, :2])
ax1 = fig.add_subplot(grid[1:, -1])
```
