```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023,]
income = [82000, 83500, 84000, 85000, 85500, 85700, 86000, 86000, 86200,]
expenses = [80000, 81000, 82000, 82500,83300, 84000, 85000, 85700, 85750]

plt.plot(years, income,'.--g', label = 'Income', linewidth = 3)
plt.plot(years, expenses, label = 'Expense', color = 'orange', linestyle = ':', marker = 'P', linewidth = 3)
plt.legend(loc='best', ncols = 2, prop = {'family': 'serif', 'size': 15, 'style': 'italic'}, edgecolor = 'black', labelcolor = 'r', fancybox = False, facecolor = 'c', title ='Legend',
          columnspacing =1, labelspacing = 2,shadow = False)
plt.show()
```

```python
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
fig = plt.figure(facecolor = 'b')
ax = fig.add_axes([0,0,0.5,0.5])
ax.plot(x,y)
ax.set_title("sine wave")
ax.set_xlabel('angle')
ax.set_ylabel('sine')

x = np.arange(0, math.pi*2, 0.05)
y = np.cos(x)
fig = plt.figure(facecolor = 'b')
ax1 = fig.add_axes([0,0,0.5,0.5])
ax1.plot(x,y)
ax1.set_title("sine wave")
ax1.set_xlabel('angle')
ax1.set_ylabel('sine')

plt.show()
```
