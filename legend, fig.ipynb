```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023,]
income = [82000, 83500, 84000, 85000, 85500, 85700, 86000, 86000, 86200,]
expenses = [80000, 81000, 82000, 82500,83300, 84000, 85000, 85700, 85750]

plt.plot(years, income,'.--g', label = 'Income', linewidth = 3)
plt.plot(years, expenses, label = 'Expense', color = 'orange', linestyle = ':', marker = 'P', linewidth = 3)
plt.legend(loc='best', ncols = 2, prop = {'family': 'serif', 'size': 15, 'style': 'italic'}, edgecolor = 'black', labelcolor = 'r', fancybox = False, facecolor = 'b', title ='Legend',
          columnspacing = 1, labelspacing =3,shadow = True)
plt.show()
```

```python
data = np.arange(0, 200, 20)
fig = plt.figure(facecolor = 'b', figsize=(8,5))

ax1 =fig.add_subplot(121)
ax1.plot(data, np.tan(data))
plt.title('tan')

ax2 =fig.add_subplot(122)
ax2.plot(data, np.sin(data))
plt.title('sin')


plt.show()
```
