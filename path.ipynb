```python
import matplotlib.pyplot as plt
from matplotlib.path import Path
import matplotlib.patches as patches
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.set_xlim(-3,3)
ax.set_ylim(-3,3)

verts = [(-2.,0.),(-1.,-2.),(2., 2.5),(-1.,1.5),(-2,0.)]
codes = [Path.MOVETO, Path.LINETO, Path.CURVE3, Path.LINETO, Path.CLOSEPOLY]
path = Path(verts, codes)

draw = patches.PathPatch(path, lw = 2)
ax.add_patch(draw)

plt.show()
```
