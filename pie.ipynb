```python
import matplotlib.pyplot as plt
```

```python
os_chart = ['Windows', 'Os x', 'Unknown', 'Linux', 'Chrom Os']
market_share = [73.5, 14.7, 5.34, 3.88, 2.56,]

plt.pie(market_share,labels=os_chart, colors =['b','r','g','c','y'],autopct='%1.1f%%', explode =[0,0.1,0.2,0.3,0.4],
radius=1, startangle =90,counterclock =False, frame =False,rotatelabels = True, shadow = False, hatch = '/')
plt.legend(os_chart, loc='best', bbox_to_anchor=(0.6, 0.6, 0.6, 0.6))
plt.show()
```
