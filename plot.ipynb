```python
import matplotlib.pyplot as plt
```

```python
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023,]
income = [82000, 83500, 84000, 85000, 85500, 85700, 86000, 86000, 86200,]
expenses = [80000, 81000, 82000, 82500,83300, 84000, 85000, 85700, 85750]

plt.plot(years, income,'o--g',label = 'Income' )
plt.plot(years,expenses, label = 'Expenses', marker = 'D', linestyle = '-.', color = 'orange')
plt.title('income vs expenses')
plt.xlabel('Years')
plt.ylabel('USD')
plt.legend()
plt.savefig('plot.png')
plt.grid()
plt.show()
```
