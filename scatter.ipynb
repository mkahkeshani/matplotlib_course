```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.rand(30)
y = np.random.rand(30)
size = np.random.rand(30)

plt.scatter(x,y, s = 100, c=size, cmap = 'Set2', alpha = 0.7, edgecolors= 'black' )
plt.title('Random')
plt.xlabel('X')
plt.ylabel('Y')
plt.colorbar(location = 'left')
m, b = np.polyfit(x,y,1)
plt.plot(x,m * x + b, color = 'red')
labels = ['X', 'Y']
plt.legend(labels)
plt.show()
```
