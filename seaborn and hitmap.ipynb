```python
import  seaborn as sns
import matplotlib.pyplot as plt
```

```python
flight_long = sns.load_dataset('flights')
flight_long
```

```python
wide = flight_long.pivot(index='year', columns='month', values='passengers')
wide
```

```python
sns.heatmap(wide, annot=True, fmt = 'd', linewidths=0.5)
```
