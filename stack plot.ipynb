```python
import matplotlib.pyplot as plt
```

```python
days = [1, 2, 3, 4, 5]

representation_of_Study = {
'studying' : [7, 8, 6, 11, 7],
'playing' :  [8, 5, 7, 8, 13],
'sleeping':[9, 11, 11, 5, 4]
}

plt.stackplot(days,representation_of_Study.values(), labels =representation_of_Study.keys(), baseline = 'wiggle')
plt.xlabel('x')
plt.ylabel('y')
plt.title('representation_of_Study')
plt.legend()
plt.show()
```
